```python
import pandas as pd
import statsmodels.api as sm
import math
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/adiacla/bigdata/master/titanic.csv")
df
```

```python
#Check nulls
df.isna().sum()
```

```python
df = df.drop('Cabin', axis =1)
```

```python
df = df.dropna()
df
```

# Model Logistic Regression

x -> Ind. 
y -> Dep. Survided

```python
df.dtypes
```

```python
x = df.iloc[:,[5,6,7]]
y = df.Survived
logit_model = sm.Logit(y,x)
result = logit_model.fit()
print(result.summary())
#Converged: Very important is true
```

```python
#Probabilidad profesor.
#logit(P(A)) = -0,0135*Edad (43) -0.1952*Sib(1)Sp + 0,2616*Parch(1)
#Calculo de probabilidad.
print("Probabilidad Profe. Feisar:",(-0.0135*43-0.1952*1+0.2616*1))
print((1/1 + math.e)**-0.5141)
print("Probabilidad José:",(-0.0135*22-0.1952*0+0.2616*0))
print((1/1 + math.e)**-0.297)
```

```python
#Taller
# 1. Agregar un análisis según el sexo
# 2. Agregar el Sex y Embarked
# Incluir en el grupo de variables independientes.
```

```python
df[["Sex","Embarked"]] = df[["Sex","Embarked"]].astype("category")
df2 = pd.get_dummies(df[['Sex','Embarked']], prefix=['Sex','Embarked'])
df2
```

```python
df.head()
```

```python
# Convertir los datos categoricos a númericos de la columna sexo
# Hombre: 1 y Mujer: 2.

df['Sex'].replace(['male', 'female'],
                        [1, 2], inplace=True)
```

```python
df.head()
```

```python
# Convertir los datos categoricos de la columna embarque ("Embarked": Port embarkation)
#(C - Cherbourg, S - Southampton, Q = Queenstown)

df['Embarked'].replace(['C', 'S', 'Q'],
                        [1, 2, 3], inplace=True)
```

```python
df.head()
```

```python
#Se confirma el cambio de los datos de la variables sexo y embarque, por valores númericos
df.dtypes
```

```python
df2.head()
```

```python
#Uso del modelo agregando las variables de sexo y Puerto de Embarque

x = df.iloc[:,[4,5,6,7,10]]
y = df.Survived
logit_model = sm.Logit(y,x)
result = logit_model.fit()
print(result.summary())
#Converged: Very important is true
```
